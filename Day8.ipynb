<h1 style="font-size:50px; color:red">Literals</h1>

Literals is a raw data given in a variable. In python, There are various types of literals they are follows:

<ul style="font-size:20px">
<li> Numericals Literals</li>
    <br/>
<li> String Literals</li>
<br/>
    <li> Boolean Literals</li>
    <br/>
<li> Special Literals</li>
</ul>

<h1 style="font-size:30px; color:blue">Numeric Literals</h1>

```python
#binary literals
a = 0b1010
a
```

```python
#decimal literals
b = 100
b
```

```python
print(b.real,b.imag)
```

```python
#Octal Literals
c = 0o310
c
```

```python
#Hexadecimal Literals
d = 0x12c
d
```

```python
#float literals
f1 = 10.5
f1
```

```python
print(f1.imag,f1.real)
```

<h1 style="font-size:30px; color:blue">String Literals</h1>

```python
s1 = 'this is Python'
s2 = "this is python"
char = "C"
multiline_str = """This is Python"""
unicode = u"\U0001F600\U0001F606\U0001F923" 
raw_str = r"raw \n string"
raw = "raw \n string"

print(s1)
print(s2)
print(char)
print(multiline_str)
print(unicode)
print(raw_str)
print(raw)
```

<h1 style="font-size:30px; color:blue">Boolean Literals</h1>

```python
a = True+4
b = False+10

print("a: ",a)
print("b: ",b)
```
