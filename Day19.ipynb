<h1 style="font-size:25px; color: #E48400">4. Arbitrary Arguments</h1>

```python
# **this is also the example of packing in python*** #
def flexi(*num):
    prod=1
    for i in num:
        prod *= i
    print(prod)
```

```python
flexi(4)
```

```python
flexi(2,3)
```

```python
flexi(5,6,7,8)
```

<h1 style="font-size:30px; color: blue">Global vs Local variable</h1>

```python
def func_a():
    print("inside func_a")

def func_b(y):
    print ("inside func_b")
    return y

print(func_a())
print(5+func_b(2))
```

```python
def f():
    def x(a, b):
        return a+b
    return x

val = f()(3,4)
print(val)
```

```python
def f(y):
    x = 1
    x += 1
    print(x)
x=5
f(x)
print(x)
```

```python
def g(y):
    print(x) 
    print(x+1)

    
x = 5
g(x)
print(x)
```

```python
def h(y):
    x+=1   # you can only use global variable inside the function but can't change
x=5
h(x)
print(x)
```

```python
# you can do but need to add one line global x
def h(y):
    global x
    x+=1   
x=5
h(x)
print(x)
```
