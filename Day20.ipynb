<h1 style="font-size:30px; color: blue">Multiplication</h1>

```python
def mul(a,b):
    if b==1:
        return a
    else:
        return a+mul(a,b-1)
```

```python
mul(2,3)
```

<h1 style="font-size:30px; color: blue">Factorial</h1>

```python
def fact(a):
    if(a==1):
        return 1
    else:
        return a*fact(a-1)
```

```python
fact(5)
```

```python
fact(1)
```

<h1 style="font-size:30px; color: blue">Palindrome</h1>

```python
def palin(text):
    if len(text)<=1:
        print("palindrome")
    else:
        if text[0] == text[-1]:
            palin(text[1:-1])
        else:
            print("Not a Palin")
```

```python
palin("madam")
palin("malayalam")
palin("python")
palin("abba")
```

<h1 style="font-size:30px; color: blue">Fibbonacci</h1>

```python
def fib(m):
    if m==1 or m==0:
        return 1
    else:
        return fib(m-1)+fib(m-2)
```

```python
fib(1)
```

```python
fib(2)
```

```python
fib(3)
```

```python
fib(4)
```

```python
fib(5)
```
